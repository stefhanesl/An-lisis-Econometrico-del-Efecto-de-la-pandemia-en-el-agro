**FUNCIONES**

Tratamiento de compra por insumos

> si ec19_actafec == 0 es grupo de control -> 0

> si ec19_actafec == 1 puede ser tratamiento o no ocuparse:

    > si ec19_actafec == 1 y ec19_dadinsu == 1 va directamente al grupo tratamiento -> 1
    > si ec19_actafec == 1 y ec19_dadinsu == 0 se ignora -> np.NaN

```python
def compra_por_insumos(fila):
  if fila['ec19_actafec'] == 0:
    return 0
  if fila['ec19_actafec'] == 1 and fila['ec19_dadinsu'] == 0:
    return np.NaN
  if fila['ec19_actafec'] == 1 and fila['ec19_dadinsu'] == 1:
    return 1
```

Tratamiento de ventas de productos por transporte

> si ec19_actafec == 0 es grupo de control -> 0

> si ec19_actafec == 1 puede ser tratamiento o no ocuparse:

    > si ec19_actafec == 1 y ec19_pveprfalt == 1 va directamente al grupo tratamiento -> 1
    > si ec19_actafec == 1 y ec19_pveprfalt == 0 se ignora -> np.NaN

```python
def ventas_por_transporte(fila):
  if fila['ec19_actafec'] == 0:
    return 0
  if fila['ec19_actafec'] == 1 and fila['ec19_pveprfalt'] == 0:
    return np.NaN
  if fila['ec19_actafec'] == 1 and fila['ec19_pveprfalt'] == 1:
    return 1
```

**tratamiento de ventas de productos por demanda**

> si ec19_actafec == 0 es grupo de control -> 0

> si ec19_actafec == 1 puede ser tratamiento o no ocuparse:

    > si ec19_actafec == 1 y ec19_pveprausc == 1 va directamente al grupo tratamiento -> 1
    > si ec19_actafec == 1 y ec19_pveprausc == 0 se ignora -> np.NaN

```python
def ventas_por_demanda(fila):
  if fila['ec19_actafec'] == 0:
    return 0
  if fila['ec19_actafec'] == 1 and fila['ec19_pveprausc'] == 0:
    return np.NaN
  if fila['ec19_actafec'] == 1 and fila['ec19_pveprausc'] == 1:
    return 1
```

### ARROZ

```python
ruta_arroz = os.path.join(ruta_iniciales, 'arroz.csv')
arroz = pd.read_csv(ruta_arroz)
arroz
```

```python
arroz['tratamiento_compra_insumos'] = arroz.apply(compra_por_insumos, axis=1)
arroz[['Identificador', 'ec19_actafec', 'ec19_dadinsu', 'tratamiento_compra_insumos']].head(10)
```

```python
arroz['tratamiento_ventas_transporte'] = arroz.apply(ventas_por_transporte, axis=1)
arroz[['Identificador', 'ec19_actafec', 'ec19_pveprfalt', 'tratamiento_ventas_transporte']].head(10)
```

```python
arroz['tratamiento_ventas_demanda'] = arroz.apply(ventas_por_demanda, axis=1)
arroz[['Identificador', 'ec19_actafec', 'ec19_pveprausc', 'tratamiento_ventas_demanda']].head(10)
```
