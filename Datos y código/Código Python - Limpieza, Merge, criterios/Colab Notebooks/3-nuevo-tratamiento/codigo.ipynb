```python
import os
import random as rd
import numpy as np
import pandas as pd
```

```python
ruta_iniciales = os.path.join(os.getcwd(), 'archivos-iniciales')
ruta_resultados = os.path.join(os.getcwd(), 'archivos-resultados')
```

```python
archivos_iniciales = os.listdir(ruta_iniciales)
archivos_iniciales
```

```python
maiz = pd.read_csv(os.path.join(ruta_iniciales, 'maiz.csv'))
arroz = pd.read_csv(os.path.join(ruta_iniciales, 'arroz.csv'))
cacao = pd.read_csv(os.path.join(ruta_iniciales, 'cacao.csv'))
```

```python
'tratamiento_ventas_transporte' in maiz.columns and 'tratamiento_ventas_demanda' in maiz.columns
```

```python
'tratamiento_ventas_transporte' in arroz.columns and 'tratamiento_ventas_demanda' in arroz.columns
```

```python
'tratamiento_ventas_transporte' in cacao.columns and 'tratamiento_ventas_demanda' in cacao.columns
```

### FUNCIONES

```python
def nuevo_tratamiento(fila):
  transporte = fila['tratamiento_ventas_transporte']
  demanda    = fila['tratamiento_ventas_demanda']
  array      = np.array([transporte, demanda])
  array      = np.nan_to_num(array, nan=0)
  resultado  = array.sum()
  if resultado > 1:
    resultado -= 1
  return resultado
```

### TRABAJANDO

```python
maiz['final_ventas'] = maiz.apply(nuevo_tratamiento, axis=1)
maiz[['tratamiento_ventas_transporte', 'tratamiento_ventas_demanda', 'final_ventas']].head(10)
```

```python
arroz['final_ventas'] = arroz.apply(nuevo_tratamiento, axis=1)
arroz[['tratamiento_ventas_transporte', 'tratamiento_ventas_demanda', 'final_ventas']].head(10)
```

```python
cacao['final_ventas'] = cacao.apply(nuevo_tratamiento, axis=1)
cacao[['tratamiento_ventas_transporte', 'tratamiento_ventas_demanda', 'final_ventas']].head(10)
```

### GUARDADO

```python
maiz.to_csv(os.path.join(ruta_resultados, 'maiz.csv'), index=False, encoding='utf-8')
maiz.to_stata(os.path.join(ruta_resultados, 'maiz.dta'), write_index=False)

arroz.to_csv(os.path.join(ruta_resultados, 'arroz.csv'), index=False, encoding='utf-8')
arroz.to_stata(os.path.join(ruta_resultados, 'arroz.dta'), write_index=False)

cacao.to_csv(os.path.join(ruta_resultados, 'cacao.csv'), index=False, encoding='utf-8')
cacao.to_stata(os.path.join(ruta_resultados, 'cacao.dta'), write_index=False)
```
