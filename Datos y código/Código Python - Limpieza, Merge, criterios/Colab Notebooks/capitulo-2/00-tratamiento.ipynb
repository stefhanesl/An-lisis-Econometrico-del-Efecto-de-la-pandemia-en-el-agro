```python
import os
import numpy as np
import pandas as pd
```

```python
originales = os.path.join(os.getcwd(), '00-archivos-originales-capitulo-1')
originales
```

```python
archivos_originales = [x for x in os.listdir(originales) if x.endswith('.csv')]
archivos_originales
```

```python
class Archivo:

  def __init__(self, directorio):
    
    self.archivo = pd.read_csv(directorio)
  
  def chequear_columnas_comun(self, otrasColumnas):
    columnas = set(self.archivo.columns)
    otrasColumnas = set(otrasColumnas)
    return columnas & otrasColumnas
  
  def tienes_las_otras_columnas(self, otrasColumnas):
    columnas = set(self.archivo.columns)
    otrasColumnas = set(otrasColumnas)
    return otrasColumnas.issubset(columnas)

  def columnas_no_hay(self, otrasColumnas):
    columnas = set(self.archivo.columns)
    otrasColumnas = set(otrasColumnas)
    return otrasColumnas - columnas
  
  def guardar_nuevo(self, nuevo_nombre, columnas):
    self.archivo[columnas].to_csv(nuevo_nombre, index=False)
```

```python
otrasColumnasCP = ['cg_k100', 'cg_k101', 'cg_k102', 'eu_k1301', 'cp_k409ha', 'cp_k411ha', 'cp_afecta_prod',	'cp_riego', 'cp_aferti', 'ventas', 'tratamiento_compra_insumos', 'tratamiento_ventas_transporte', 'tratamiento_ventas_demanda', 'cp_prod', 'cp_cantidad_npk_fq']
otrasColumnasCT = ['cg_k100', 'cg_k101', 'cg_k102', 'eu_k1301', 'ct_k509', 'ct_k511ha', 'ct_afecta_prod',	'ct_riego', 'ct_aferti', 'Ventas', 'tratamiento_compra_insumos', 'tratamiento_ventas_transporte', 'tratamiento_ventas_demanda', 'ct_prod', 'ct_cantidad_npk_fq']
```

```python
for nombre_archivo in archivos_originales:
  directorio = os.path.join(originales, nombre_archivo)
  archivo = Archivo(directorio)

  if 'cp' in nombre_archivo.split('-'):
    otrasColumnas = otrasColumnasCP
  else:
    otrasColumnas = otrasColumnasCT

  print('-'*45)
  print(nombre_archivo)
  print()

  print(archivo.chequear_columnas_comun(otrasColumnas))
  print()

  print(archivo.tienes_las_otras_columnas(otrasColumnas))
  print()

  print(archivo.columnas_no_hay(otrasColumnas))
  print()

  if archivo.tienes_las_otras_columnas(otrasColumnas):
    nuevo_nombre = '-'.join(nombre_archivo.split('-')[2:])
    columnas_tomar = ['Identificador', 'ual_parr', 'pa_arealm', 'pa_areempa', 'eq_earado', 'eq_semman', 'eq_emotoguadana'] + otrasColumnas
    archivo.guardar_nuevo(nuevo_nombre, columnas_tomar)
```
