```python
import os
import random as rd
import numpy as np
import pandas as pd
```

```python
dir_archivos = os.path.join(os.getcwd(), '01-archivos-nuevos-capitulo-2')
dir_archivos
```

```python
archivos = [x for x in os.listdir(dir_archivos) if x.endswith('.csv')]
archivos.sort()
archivos
```

```python
class ArchivoSinIdRepetidos:

  def __init__(self, directorio):
    self.df = pd.read_csv(directorio)
    self.df_unicos = self.df.drop_duplicates(subset = ['Identificador'])

  def tratamiento_columnas_sumar(self, columnas_sumar):
    self.df_unicos.set_index('Identificador', inplace=True)
    for identificador in np.unique(self.df['Identificador']):
      for columna in columnas_sumar:
        suma = self.df[self.df['Identificador'] == identificador][columna].sum()
        self.df_unicos.loc[identificador, columna] = suma
    self.df_unicos.reset_index(inplace=True)
  
  def tratamiento_columna_riego(self, columna_riego):
    self.df_unicos.set_index('Identificador', inplace=True)

    for identificador in self.df_unicos.index:
      datos_riego = self.df[self.df['Identificador'] == identificador][columna_riego]
      if datos_riego.nunique() > 1:
        conteo = datos_riego.value_counts().to_dict()
        if conteo[0] > conteo[1]:
          self.df_unicos.loc[identificador, columna_riego] = 0
        else:
          self.df_unicos.loc[identificador, columna_riego] = 1
    
    self.df_unicos.reset_index(inplace=True)
      
  def guardar_nuevo(self, nombre):
    self.df_unicos.to_csv(nombre, index=False)
```

```python
columnas_sumar_cp = ['cp_k409ha', 'cp_k411ha', 'ventas', 'cp_prod', 'cp_cantidad_npk_fq']
columnas_sumar_ct = ['ct_k509', 'ct_k511ha', 'Ventas', 'ct_prod', 'ct_cantidad_npk_fq']
for nombre_archivo in archivos:
  columnas_sumar = columnas_sumar_cp if nombre_archivo.startswith('cp') else columnas_sumar_ct
  columna_riego = 'cp_riego' if nombre_archivo.startswith('cp') else 'ct_riego'
  archivo = ArchivoSinIdRepetidos(os.path.join(dir_archivos, nombre_archivo))
  archivo.tratamiento_columnas_sumar(columnas_sumar)
  archivo.tratamiento_columna_riego(columna_riego)
  archivo.guardar_nuevo(nombre_archivo)
```
