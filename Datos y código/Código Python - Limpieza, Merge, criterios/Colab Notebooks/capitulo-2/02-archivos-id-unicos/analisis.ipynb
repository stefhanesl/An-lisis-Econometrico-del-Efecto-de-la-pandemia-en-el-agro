```python
from scipy.stats import ttest_ind
```

```python
bool0 = df['tratamiento_compra_insumos'] == 0
bool1 = df['tratamiento_compra_insumos'] == 1
```

```python
for col in df.columns[1:]:
  x = df[col][bool0]
  y = df[col][bool1]
  print(col)
  print(x.mean())
  print(y.mean())
  print(ttest_ind(x, y))
  print()
```
