```python
class ArchivoOriginalSucio:

  def __init__(self, path):
    self.path = path
    self.data = pd.read_csv(self.path, sep=';')
  
  def tipo_datos_columnas(self):
    return self.data.dtypes

  def tiene_ids_repetidos(self):
    return self.data['Identificador'].duplicated().sum()

  def quitar_espacios_blanco(self):
    self.data = self.data.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    self.data = self.data.replace('', np.NaN)
    print('Se han quitado los espacios en blanco')
  
  def modificar_sep_decimales(self):
    self.data = self.data.applymap(lambda x: x.replace(',', '.') if isinstance(x, str) else x)
    print('Se ha establecido el punto como separador de decimales')

  
  def guardar_archivo(self, ruta_guardar):
    self.data.to_csv(ruta_guardar, sep=',', index=False, encoding='utf-8')
    print(f'Archivo guardado en {ruta_guardar}')
```

```python
path = os.getcwd()
archivos = os.listdir(path + '/originales-sucios')
archivos
```

**Limpiar covid**

```python
covid = ArchivoOriginalSucio('originales-sucios/' + archivos[0])
covid.data
```

```python
covid.tipo_datos_columnas()
```

```python
covid.tiene_ids_repetidos()
```

```python
covid.quitar_espacios_blanco()
```

```python
covid.data.head()
```

```python
covid.guardar_archivo('datos-covid-limpios.csv')
```
