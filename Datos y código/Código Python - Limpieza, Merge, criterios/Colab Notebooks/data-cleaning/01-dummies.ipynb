```python
path = os.getcwd()
archivos = os.listdir(path + '/01-originales-limpios')
archivos
```

```python
def convertir_columnas_dummies(df):
  for column in df.columns:
    if df[column].nunique() == 2:
      print(column)
      print(df[column].value_counts())
      df[column] = pd.Categorical(df[column])
      df[column] = df[column].apply(lambda x: 0 if x == 2 else x)
      print(df[column].value_counts())
      print('-'*33)
  return df
```

**Trabajamos con COVID**

```python
covid = pd.read_csv('01-originales-limpios/' + archivos[0])
covid
```

```python
covid['ec19_recayugadm'].value_counts()
```

```python
covid.dtypes
```

```python
covid = convertir_columnas_dummies(covid)
```

```python
covid.dtypes
```

```python
covid['ec19_recayugadm'].value_counts()
```

```python
covid.to_csv('datos-covid.csv', sep=',', encoding='utf-8', index=False)
```
