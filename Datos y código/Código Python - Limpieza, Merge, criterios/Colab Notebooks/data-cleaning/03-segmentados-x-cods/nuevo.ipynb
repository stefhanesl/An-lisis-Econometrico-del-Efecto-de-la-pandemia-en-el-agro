**trabajamos maiz**

```python
maiz = pd.read_csv('ct-maiz.csv')
maiz.sort_values('Identificador', inplace=True)
maiz
```

```python
def addForwardZero(numero):
  while len(numero) < 2:
    numero = '0' + numero
  return numero
```

```python
maiz['ct_numt'] = maiz['ct_numt'].astype('str')
maiz['ct_numt'].dtype
```

```python
maiz['ct_numt'] = maiz['ct_numt'].apply(addForwardZero)
maiz[['Identificador', 'ct_numt']].head(15)
```

```python
maiz['Identificador'] = maiz['Identificador'].astype('str')
maiz['Identificador'].dtype
```

```python
id = maiz['Identificador'] + maiz['ct_numt']
id
```

```python
maiz.insert(1, 'Id', id)
maiz.head(15)
```

```python
repetido = maiz.duplicated(['Identificador'])
repetido
```

```python
maiz['repetido'] = 1
maiz['repetido']
```

```python
maiz['repetido'][repetido == True] = 0
maiz['repetido']
```

```python
maiz[['Identificador', 'Id', 'ct_numt', 'repetido']].head(20)
```

```python
maiz_sin_repetidos = maiz[maiz['repetido'] == 1]
maiz_sin_repetidos[['Identificador', 'Id', 'ct_numt', 'repetido']].head(20)
```

```python
maiz_sin_repetidos.to_csv('maiz-sin-repetidos.csv', index=False)
```
