```python
principal = pd.read_csv('datos-principales.csv')
principal
```

```python
def columnas_iguales(df1, df2):
  columnas1 = set(df1.columns)
  columnas2 = set(df2.columns)
  iguales = columnas1 & columnas2
  return iguales
```

```python
def ids_iguales(df1, df2):
  ids1 = set(df1['Identificador'])
  ids2 = set(df2['Identificador'])
  ids = ids1 & ids2
  return list(ids)
```

```python
archivos = os.listdir(os.getcwd())
archivos
```

```python
archivos_cultivos = [x for x in archivos if x.startswith('cp') or x.startswith('ct')]
archivos_cultivos = archivos_cultivos[-2:-1]
archivos_cultivos
```

```python
for nombre_archivo in archivos_cultivos:
  tipo_cultivo = nombre_archivo.split('.')[0]
  # ct-arroz.csv -> ct-arroz -> esto es para cuando guarde el archivo del merge ponerle nombre
  df_cultivo = pd.read_csv(nombre_archivo)
  columnas = columnas_iguales(principal, df_cultivo) - {'fact_exp_fin'}
  # me interesan todas las columnas que se comparten menos fact_exp_fin ya que hay valores distintos en el archivo de datos principales y el de cultivos
  columnas = list(columnas)
  df = pd.merge(principal, df_cultivo, how='inner', on=columnas, suffixes=('_principales', '_cultivo'))
  # el inner es para que tome los identifadores que se estan tanto en el cultivo como en el de datos principales
  df.to_csv(f'data-{tipo_cultivo}.csv', sep=',', encoding='utf-8', index=False)
```
