```python
import os
import numpy as np
import pandas as pd
```

```python
def convertir_columna_categorica(df):
  for columna in df.columns:
    if df[columna].nunique() == 2:
      df[columna] = pd.Categorical(df[columna])
  return df
```

```python
covid = pd.read_csv('datos-covid.csv')
covid = convertir_columna_categorica(covid)
covid
```

**tratamiento de compra de insumos**

```python
# tratamiento_compra_insumos

# si ec19_actafec == 0 es grupo de control -> 0

# si ec19_actafec == 1 puede ser tratamiento o no ocuparse:

  # si ec19_actafec == 1 y ec19_dadinsu == 1 va directamente al grupo tratamiento -> 1
  # si ec19_actafec == 1 y ec19_dadinsu == 0 se ignora -> np.NaN
```

```python
def determinar_cat1(fila):
  if fila['ec19_actafec'] == 0:
    return 0
  if fila['ec19_actafec'] == 1 and fila['ec19_dadinsu'] == 0:
    return np.NaN
  if fila['ec19_actafec'] == 1 and fila['ec19_dadinsu'] == 1:
    return 1
```

```python
covid['tratamiento_compra_insumos'] = covid.apply(determinar_cat1, axis=1)
```

```python
covid[['Identificador', 'ec19_actafec', 'ec19_dadinsu', 'tratamiento_compra_insumos']].head(10)
```

```python
def columnas_iguales(df1, df2):
  columnas1 = set(df1.columns)
  columnas2 = set(df2.columns)
  iguales = columnas1 & columnas2
  return iguales
```

```python
columnas_iguales(covid, empleo)
```

```python
columnas_iguales(covid, caracteristicas)
```

```python
columnas_iguales(empleo, caracteristicas)
```

```python
def ids_iguales(df1, df2):
  ids1 = set(df1['Identificador'])
  ids2 = set(df2['Identificador'])
  resultado = ids1 & ids2
  print(len(resultado))
```

```python
ids_iguales(covid, empleo)
```

```python
ids_iguales(covid, caracteristicas)
```

```python
ids_iguales(empleo, caracteristicas)
```

**Vamos a hacer el merge**

```python
columnas = list(columnas_iguales(covid, empleo))
df = pd.merge(covid, empleo, how='left', on=columnas)
df
```
