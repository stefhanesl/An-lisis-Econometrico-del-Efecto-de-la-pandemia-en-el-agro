```python
import os
import numpy as np
import pandas as pd
```

```python
archivos = os.listdir()
archivos
```

```python
archivos_csv = [x for x in archivos if x.endswith('.csv')]
archivos_csv
```

```python
for nombre_archivo in archivos_csv:
  nombre = nombre_archivo.split('.')[0]
  nombre_stata = nombre + '.dta'
  df = pd.read_csv(nombre_archivo)
  df.to_stata(nombre_stata, write_index=False)
```
